```python
df0=pd.read_csv('real time stock data.csv',parse_dates=True,infer_datetime_format='%Y-%m-%d %H:%M:%S',index_col=1)
```

```python
df =df0.copy()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
upsampled=df.resample('S')
```

```python
interpolated = upsampled.interpolate(method='linear')
```
